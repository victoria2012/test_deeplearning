```python
posts = df['Title']+['Description']
type(posts)
```

```python
from konlpy.tag import Okt
tagger = Okt()
```

```python
stop_words = ['강,', '글,', '애', '미', '번', '은', '이', '것', '등', '더', '를', '좀', '즉', '인', '옹', '때', '만', '원', '이때', '개']
```

```python
posts_noun = []
for post in posts:               # 빈도를 측정하는 과정임. 명사만 꺼낸다.  
  # print(tagger.nouns(post))
  for noun in tagger.nouns(post):   # 불용어를 삭제하는 작업임(여기서는 생략, 이후에 스스로 해볼 것)
    posts_noun.append(noun)  

len(posts_noun), type(posts_noun)
```

```python
posts_noun
```

```python
from collections import Counter
```

```python
noun_counter = Counter(posts_noun)
type(noun_counter)
```

```python
top_30_noun = noun_counter.most_common(30)
type(top_30_noun), top_30_noun
```

```python
top_nouns_dict = dict(top_30_noun)
type(top_nouns_dict)
```

```python
from wordcloud import WordCloud
```

```python
path = ''
```

```python
nouns_wordcloud = WordCloud()
nouns_wordcloud.generate_from_frequencies(top_nouns_dict)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(nouns_wordcloud)
```
